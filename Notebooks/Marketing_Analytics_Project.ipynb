## Context
To help the marketing team **optimize advertising spend** by analyzing user behavior, sales patterns, and marketing performance.

## Objective
The goal of this project is to answer key business questions, including:
- How do customers use the service?
- When do they start making purchases?
- How much revenue does each customer generate?
- At what point does the revenue cover the cost of customer acquisition (CAC)?
- Which marketing channels deliver the highest return on marketing investment (ROMI)?

## Data Sources
The dataset includes:
1. **Visits data** – server logs with user sessions (January 2017 – December 2018).
2. **Orders data** – customer purchases with revenue details.
3. **Marketing costs data** – advertising spend per channel over time.

1. **Data Preparation** – Load and clean data, ensure correct data types.
2. **Analysis**:
   - Calculate visits, sessions, and retention rates.
   - Analyze sales patterns, average order value, and customer lifetime value (LTV).
   - Evaluate marketing spend, CAC, and ROMI by source.
3. **Visualization** – Create graphs to compare metrics by device, channel, and over time.
4. **Recommendations** – Suggest where and how much to invest in marketing based on data-driven insights.

### Number of Users Who Use the Product Daily, Weekly, and Monthly

#### Session Duration

To calculate the average session duration, we subtract the start time (start_ts) from the end time (end_ts) for each record in the visits dataset. This will give us the session length for each visit, which we can then analyze to find average, median, and distribution patterns.

```python
# Ensure datetime format
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])

# Calculate session duration in minutes
visits['session_duration'] = (visits['end_ts'] - visits['start_ts']).dt.total_seconds() / 60

# Summary statistics
visits['session_duration'].describe()
```

#### When do people start buying? (Conversion Time) :

```python
# 1) Ensure datetime types
orders['buy_ts']  = pd.to_datetime(orders['buy_ts'], errors='coerce')
visits['start_ts'] = pd.to_datetime(visits['start_ts'], errors='coerce')

# 2) First touchpoints (first visit and first order per user)
first_visits = (visits
                .groupby('uid', as_index=False)['start_ts']
                .min()
                .rename(columns={'start_ts':'first_visit_ts'}))

first_orders = (orders
                .groupby('uid', as_index=False)['buy_ts']
                .min()
                .rename(columns={'buy_ts':'first_order_ts'}))

# 3) Join and compute conversion time (days)
conversion = (first_visits
              .merge(first_orders, on='uid', how='left'))

# Difference in whole days (NaT -> NaN)
conversion['time_to_convert_days'] = (conversion['first_order_ts'] - conversion['first_visit_ts']).dt.days

# 4) Monthly cohort based on first visit
conversion['cohort_month'] = conversion['first_visit_ts'].dt.to_period('M').dt.start_time

# 5) Buckets: 0d (same day), 1d (next day), 2d+ (2 or more days)
conversion['conv_0d']  = (conversion['time_to_convert_days'] == 0)
conversion['conv_1d']  = (conversion['time_to_convert_days'] == 1)
conversion['conv_2dp'] = (conversion['time_to_convert_days'] >= 2)

# 6) Cohort-level rates (share of all users in the cohort)
cohort_conv = (conversion
               .groupby('cohort_month', as_index=False)[['conv_0d','conv_1d','conv_2dp']]
               .mean()
               .rename(columns={'conv_0d':'Conversion 0d',
                                'conv_1d':'Conversion 1d',
                                'conv_2dp':'Conversion 2d+'}))

# Format as percentages for display (optional)
cohort_conv_pct = cohort_conv.copy()
for col in ['Conversion 0d','Conversion 1d','Conversion 2d+']:
    cohort_conv_pct[col] = (cohort_conv_pct[col] * 100).round(2)

cohort_conv_pct.sort_values('cohort_month').head(12)
```

The pattern shows more immediate purchase behavior in recent cohorts.
The ability to recover users who don’t purchase on their first visit has weakened.

Suggested actions:

- Optimize the funnel for first-time visitors (landing pages, CTAs, upfront discounts).

- Strengthen longer-term retargeting campaigns (beyond the first 24 hours) to regain some delayed conversions.

### How Much Revenue Do Customers Bring? (LTV)

```python
# --- Build cohort revenue and sizes ---

# Revenue by (cohort month of first order) x (order month)
cohorts2 = (
    orders1
    .groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)

# Number of customers in each cohort (unique buyers in the first-order month)
cohort_sizes = (
    first_orders
    .groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
    .rename(columns={'uid': 'n_clients'})
)
```

```python
# Combine sizes with revenue
report = pd.merge(cohort_sizes, cohorts2, on='first_order_month')

report.head()
```

```python
# --- Compute cohort age (in months) and LTV per age ---
report['first_order_month'] = report['first_order_month'].astype('datetime64[ns]')
report['age'] = ((report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')).round().astype(int)

# Gross profit proxy = revenue (dataset has revenue net to platform)
report['gp']  = report['revenue']

# LTV for a cohort at a given age = revenue / number of clients in that cohort
report['ltv'] = report['gp'] / report['n_clients']
```

```python
# --- Pivot to LTV by cohort age and make it cumulative across ages ---
ltv_by_age = report.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
).sort_index()

# Cumulative LTV across months since first purchase
ltv_cum = ltv_by_age.cumsum(axis=1)

# Pretty index for plotting
ltv_cum.index = pd.to_datetime(ltv_cum.index).strftime('%Y-%m-%d')
ltv_by_age.fillna('')
```

```python
# --- Heatmap ---
plt.figure(figsize=(14, 8))
sns.heatmap(ltv_cum, annot=True, fmt='.2f', cmap='plasma')
plt.title('Cumulative LTV by Cohort and Age (Months)')
plt.xlabel('Cohort Age (Months Since First Order)')
plt.ylabel('Cohort (First Order Month)')
plt.tight_layout()
plt.show()
```

Sey Observations
Highest LTV: The cohort from September 2017 shows the strongest performance, reaching a cumulative LTV of 13.44 by month 8 — the brightest yellow square on the chart.

Early strong performers: July 2017 and June 2017 cohorts also reach relatively high LTVs (~11.88 by month 11–12).

Recent cohorts: Later cohorts (2018) have lower cumulative LTVs, partly because they haven’t had as much time to mature and accumulate revenue.

Growth pattern: Most cohorts show a steady increase in LTV over the first months, with some flattening after month 7–8, meaning revenue per customer tends to stabilize.

```python
# Calculating ROMI
romi = (orders_with_user_firsts
        .groupby(['first_order_month','source_id'])
        .agg(total_rev = ('revenue','sum'),n_buyers=('uid','nunique'))
        .reset_index()
       )
romi['ltv'] = romi['total_rev']/romi['n_buyers']
romi = pd.merge(romi, cac_avg, on='source_id')
romi['romi'] = romi['ltv']/romi['cac_avg']
pd.pivot_table(romi, values='romi', index='first_order_month', columns='source_id') 
```
